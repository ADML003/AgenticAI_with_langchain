```python
import os 
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY") # set your API key in the environment variable
model=init_chat_model(model="llama-3.3-70b-versatile", model_provider="groq", temperature=0.7)
```

```python
model.invoke("What is the capital of France?")
```

text prompts

are strings which are ideal for straight forward generation tasks without the need of retaining conversation history 


```python
model.invoke("what is uv package manager")
```

<h2>Message prompts</h2>

list of messages to the model 

system message - how model should behave
ai message - responses generated by the model 
human message - user input and interactions with the model
tool message - output of the tools

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

mess=[
    SystemMessage(content="You are a Cricket Expert."),
    HumanMessage("Who won the 2007 T20 World Cup?"),
    
    
]

res=model.invoke(mess)
res.content
```

```python
res.text
```

```python
# detailed info  to llm

sys=SystemMessage("You are a Cricket Expert. You have detailed information about cricket history, players, matches, and statistics. You can provide insights, answer questions, and discuss various aspects of cricket with depth and accuracy.")

msg=[
    sys,
    HumanMessage(content="Give me your verdict of the controversial 2019 World Cup final between England and New Zealand?")]
    
res=model.invoke(msg)

res.content
```

```python
## checking how llm understand the conversation history 

ai_msg=AIMessage("i would be happy to assist you with your cricket-related questions. Please feel free to ask me anything about cricket history, players, matches, or statistics, and I'll do my best to provide you with accurate and insightful information.")

msg=[
    SystemMessage("You are a Cricket Expert. You have detailed information about cricket history, players, matches, and statistics. You can provide insights, answer questions, and discuss various aspects of cricket with depth and accuracy."),
    HumanMessage("answers my cricket-related questions"),
    ai_msg, # as if the ai responded with this message in the conversation
    
    HumanMessage("who is better between Virat and Rohit in t20 cricket?")
]

res=model.invoke(msg)
res.text
```
