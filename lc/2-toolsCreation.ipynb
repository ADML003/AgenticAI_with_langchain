<h1>Tools creation</h1>
Models can request to call tools that perform various secondary tasks like fetching data from database, searching web etc.

```python
import os 
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY") # set your API key in the environment variable
model=init_chat_model(model="llama-3.3-70b-versatile", model_provider="groq", temperature=0.7)
res=model.invoke("who win in Ford vs Ferrari?")
res.text  
```

```python
from langchain.tools import tool
@tool
def get_weather(loc:str)-> str:
    """ get the weather of a location """
    return f"its sunny in {loc}"



model_with_tools=model.bind_tools([get_weather])    
```

```python
res=model_with_tools.invoke("what is the weather in Paris?")
print(res)
```
