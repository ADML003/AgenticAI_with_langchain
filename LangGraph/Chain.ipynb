```python
from langchain_core.messages import HumanMessage, AIMessage
```

```python
from pprint import pprint

mess=[AIMessage(content=f"please tell me how ca i help you", name="LLMModel ")]

mess.append(HumanMessage(content=f"i want to learn coding in python", name="User"))

mess.append(AIMessage(content=f"Sure! I can help you with that. What specific topics in Python are you interested in learning?", name="LLMModel "))

mess.append(HumanMessage(content=f"i am looking to prepare for Django", name="User"))

for m in mess:
    m.pretty_print()
```

appending part is being taken care of by Reducers in langgraph

```python
from langchain_groq import ChatGroq
```

```python
llm=ChatGroq(model="llama-3.3-70b-versatile")



mess.append(llm.invoke(mess))
```

```python
for m in mess:
    m.pretty_print()
```

```python
res=llm.invoke(mess)
```

```python
res.response_metadata
```

```python
def add(a:int, b:int) -> int:
    
    """ 
    Add a and b
    
    return int
    
    
    """
    
    return a+b
```

```python
llm
```

```python
llm_t=llm.bind_tools([add])

tool_call=llm_t.invoke([HumanMessage(content="What is 2 + 2?",name="user")])
```
